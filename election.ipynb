```python
df = df[df.year == 2020]
df = df.loc[~df.state_po.isin(['HI', 'AK'])]
df = df.loc[df.party.isin(['DEMOCRAT', 'REPUBLICAN'])]
df = df.reset_index(drop=True)
df.head()
```

```python
df.loc[df.state_po=='DC', 'county_fips'] = 11001
df = df.dropna(subset=['county_fips'])
df.county_fips = df.county_fips.apply(lambda x: int(x))
df.county_fips = df.county_fips.apply(lambda x: str(x))
df.county_fips = df.county_fips.apply(lambda x: '0' + x if len(x)==4 else x)
df.head()
```

```python
df.state = df.state.str.capitalize()
df.county_name = df.county_name.str.capitalize()
df.party = df.party.str.capitalize()
df.head()
```

```python
len(df.state_po.value_counts())
```

```python
total = df.loc[df['mode']=='TOTAL']
total = total[['year', 'state', 'state_po', 'county_name', 'county_fips', 'party', 'candidatevotes', 'totalvotes']]
total.head()
```

```python
not_total = df.loc[df['mode']!='TOTAL']
not_total.head()
```

```python
fips_diff = set(not_total.county_fips.values.tolist()) - set(total.county_fips.values.tolist())
```

```python
non_total = df.loc[df.county_fips.isin(fips_diff)]
non_total_df = pd.DataFrame(non_total.groupby(['year', 'state', 'state_po', 'county_name', 'county_fips', 'party', 'totalvotes'])['candidatevotes'].sum().reset_index())
non_total_df.head()
```

```python
df_merge = pd.concat([total, non_total_df], axis=0)
df_merge.head()
```

```python
df_merge['pctvotes'] = df_merge.candidatevotes / df_merge.totalvotes
df_merge.head()
```

```python
merge = df_merge.pivot_table(index=['year', 'state', 'state_po', 'county_name', 'county_fips', 'totalvotes'], columns='party', values=['candidatevotes', 'pctvotes']).reset_index()
merge.head()
```

```python
merge['margin'] = merge['candidatevotes', 'Democrat'] - merge['candidatevotes', 'Republican']
merge.head()
```

```python
merge['winning'] = merge['margin'].apply(lambda x: 'D' if x > 0 else 'R')
merge.head()
```

```python
merge['pct_margin'] = abs(merge['margin'] / merge['totalvotes'])
merge.head()
```

```python
merge['pct_rep'] = round(merge['pctvotes', 'Republican'] * 100, 2)
merge['pct_dem'] = round(merge['pctvotes', 'Democrat'] * 100, 2)
merge['pct_margin'] = round(merge['pct_margin'] * 100, 2)
merge = merge.drop(columns=[('pctvotes', 'Republican'), ('pctvotes', 'Democrat')])
```

```python
merge.head()
```

```python
merge['swing'] = merge.pct_margin.apply(lambda x: 1 if x <= 9.0 else 0)
merge.head()
```

```python
merge = merge.rename({'county_fips': 'GEOID', 'totalvotes': 'votes_total'}, axis=1)
merge = merge.drop(columns=['year', 'state', 'county_name'])
merge.head()
```

```python
merge.columns.tolist()
```

```python
merge.columns = ['state', 'GEOID', 'votes_total', 'votes_dem', 'votes_rep', 'margin', 'winning', 'pct_margin', 'pct_rep','pct_dem', 'swing']
merge['GEOID'] = merge['GEOID'].astype(str)
merge.head()
```
